# **Theory Questions**

**Q1.What is unsupervised learning in the context of machine learning ?**

**Answer :-** Unsupervised learning is a type of machine learning where the model is trained on data that has no labels. The system tries to find hidden patterns, relationships, or structures in the data on its own. Common examples include clustering (like grouping similar customers) and dimensionality reduction (like simplifying data while keeping important information).

**Q2.How does K-Means clustering algorithm work?**

**Answer** K-Means clustering works by dividing data into K groups (clusters) based on similarity. It follows these steps:

Choose the number of clusters K.

Randomly select K points as initial cluster centers (called centroids).

Assign each data point to the nearest centroid.

Recalculate the centroids by taking the average of all points in each cluster.

Repeat steps 3 and 4 until the centroids stop changing or only change slightly.

**Q3.Explain the concept of a dendrogram in hierarchical clustering.**

**Answer -** A dendrogram is a tree-like diagram used in hierarchical clustering to show how data points are grouped together step by step. It starts with each data point as its own cluster and then merges the closest clusters one by one. The height of the branches shows how similar or different the clusters are — shorter branches mean more similarity. By cutting the dendrogram at a certain height, you can decide how many clusters to form.

**Q4.What is the main difference between K-Means and Hierarchical Clustering?**

**Answer -** The main difference is that K-Means requires you to predefine the number of clusters (K) before running the algorithm, while Hierarchical Clustering does not require specifying K in advance — it builds a hierarchy of clusters that you can cut at any level to form groups.


K-Means → needs K beforehand, works on flat clusters.

Hierarchical Clustering → builds a tree of clusters automatically.

**Q5.What are the advantages of DBSCAN over K-Means?**

**Answer -** DBSCAN has several advantages over K-Means:

No need to specify number of clusters in advance.

Can find clusters of any shape, not just spherical ones.

Can detect noise and outliers, labeling them as separate points.

Works well with uneven cluster sizes and densities, where K-Means often fails.

**Q6.When would you use Silhouette Score in clustering?**

**Answer -** You use the Silhouette Score to measure how well the data points fit within their assigned clusters. It helps evaluate the quality and validity of clustering results.

A high Silhouette Score (close to 1) means points are well-matched to their own cluster and far from other clusters. A low or negative score means poor clustering. It’s especially useful for comparing different numbers of clusters (like choosing the best K in K-Means).

**Q7.What are the limitations of Hierarchical Clustering?**

**Answer -** The main limitations of Hierarchical Clustering are:

 - Not suitable for very large datasets because it has high computational cost.

 - Once merged or split, clusters cannot be changed, which may lead to poor results.

 - Sensitive to noise and outliers, which can distort the cluster structure.

 - Choosing the right cut-off point in the dendrogram can be difficult.

**Q8.Why is feature scaling important in clustering algorithms like K-Means?**

**Answer -** Feature scaling is important in K-Means because the algorithm uses distance (like Euclidean distance) to form clusters. If features have different scales (for example, age in years and income in thousands), the larger-scaled feature will dominate the distance calculation. Scaling makes all features contribute equally, leading to more accurate and balanced clustering results.

**Q9. How does DBSCAN identify noise points?**

**Answer -** DBSCAN identifies noise points as data points that do not belong to any cluster. A point is considered noise if:

 - It has fewer neighboring points than the minimum required (minPts) within a given distance (eps).

 - It is not reachable from any core point (a point with enough neighbors).

These noise points are labeled as outliers and excluded from clusters.

**Q10. Define inertia in the context of K-Means?**

**Answer -** Inertia in K-Means is a measure of how well the data points fit within their clusters. It is calculated as the sum of squared distances between each data point and its cluster’s centroid.

Lower inertia means the clusters are tighter and the model fits the data better. However, very low inertia may also indicate overfitting if too many clusters are used.

**Q11. What is the elbow method in K-Means clustering?**

**Answer -** The elbow method is a technique used to find the optimal number of clusters (K) in K-Means.

You run K-Means for different K values and plot inertia (sum of squared distances) against K. As K increases, inertia decreases, but after a certain point, the improvement becomes small and the curve forms an “elbow” shape.
The point where the curve bends (the elbow) is considered the best K, balancing accuracy and simplicity.

**Q12. Describe the concept of "density" in DBSCAN.**

**Answer -** In DBSCAN, density refers to how closely packed the data points are in a region. A region is considered dense if there are many points within a certain distance (eps) from a given point.

DBSCAN forms clusters by connecting points in these dense regions. Points in low-density areas that don’t have enough nearby neighbors are treated as noise or outliers.

**Q13. Can hierarchical clustering be used on categorical data?**

**Answer -** Yes, hierarchical clustering can be used on categorical data, but not directly with regular distance measures like Euclidean distance. Instead, you must use suitable distance metrics such as Hamming distance or Jaccard similarity, which work for categorical variables.

These metrics allow hierarchical clustering to measure similarity between categories and form meaningful clusters.

**Q14. What does a negative Silhouette Score indicate?**

**Answer -** A negative Silhouette Score means that a data point is likely assigned to the wrong cluster. It is closer to points in another cluster than to points in its own cluster.

In simple terms, a negative score shows poor clustering quality and overlapping or misclassified clusters.

**Q15. Explain the term "linkage criteria" in hierarchical clustering?**

**Answer -** In hierarchical clustering, linkage criteria decide how the distance between two clusters is measured when they are being merged.

Common types include:

 - Single linkage: distance between the closest points of two clusters.

 - Complete linkage: distance between the farthest points.

 - Average linkage: average distance between all points of the two clusters.

**Q16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?**

**Answer -** K-Means may perform poorly on data with varying cluster sizes or densities because it assumes that all clusters are spherical, evenly sized, and have similar density.

If clusters have different shapes, sizes, or densities, K-Means can:

 - Misclassify points from smaller or less dense clusters.

 - Pull centroids toward larger or denser clusters.

 - Fail to detect irregular or uneven cluster boundaries.

**Q17. What are the core parameters in DBSCAN, and how do they influence clustering?**

**Answer -** The two core parameters in DBSCAN are eps (epsilon) and minPts (minimum points).

 - eps defines the maximum distance between two points to be considered neighbors. A larger eps creates fewer, bigger clusters; a smaller one creates more, smaller clusters.

 - minPts defines the minimum number of points needed to form a dense region. Higher minPts makes the algorithm stricter, labeling more points as noise.

Together, they control how DBSCAN identifies dense areas and separates clusters from noise

**Q18. How does K-Means++ improve upon standard K-Means initialization?**

**Answer -** K-Means++ improves the standard K-Means initialization by choosing better starting centroids to reduce the chances of poor clustering.

It works by:

 - Selecting the first centroid randomly.

 - Choosing the next centroids based on how far points are from existing centroids — points farther away have a higher chance of being selected.

 - Repeating until K centroids are chosen.

This leads to faster convergence and better, more stable clustering results compared to random initialization

**Q19. What is agglomerative clustering?**

**Answer -** Agglomerative clustering is a type of hierarchical clustering that builds clusters from the bottom up.

It starts with each data point as its own cluster and then repeatedly merges the two closest clusters based on a distance measure and linkage criterion. This process continues until all points are combined into one large cluster or until a desired number of clusters is reached.

**Q20. What makes Silhouette Score a better metric than just inertia for model evaluation?**

**Answer -** The Silhouette Score is better than inertia because it considers both cohesion (how close points are within a cluster) and separation (how far clusters are from each other).

Inertia only measures how tight the clusters are but ignores how well they are separated.
Silhouette Score gives a more complete evaluation by showing how well each point fits into its cluster compared to other clusters, making it a more reliable metric for clustering quality.

**Q21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot.**

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic data with 4 centers
X, y = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=1.0)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualize clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering with 4 Centers')
plt.legend()
plt.show()
```

**Q29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.**

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate synthetic data
X, y = make_blobs(n_samples=400, centers=4, random_state=42)

# Evaluate silhouette scores for k = 2 to 5
scores = []
k_values = [2, 3, 4, 5]

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

# Plot as bar chart
plt.bar(k_values, scores, color='skyblue')
plt.title("Silhouette Scores for Different K Values")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.show()
```

**Q31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries.**

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Generate overlapping data
X, y = make_blobs(
    n_samples=400,
    centers=3,
    cluster_std=2.0,  # large std → overlapping clusters
    random_state=42
)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Create a grid for decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                     np.arange(y_min, y_max, 0.05))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundaries and clusters
plt.contourf(xx, yy, Z, cmap='Pastel1', alpha=0.7)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering with Decision Boundaries (Overlapping Data)")
plt.legend()
plt.show()
```

**Q34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
line plot.**

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Compute inertia for K = 2 to 6
k_values = [2, 3, 4, 5, 6]
inertias = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

# Plot results
plt.plot(k_values, inertias, marker='o', color='blue')
plt.title("K-Means Inertia for Different K Values")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()
```

```python
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt

# Generate noisy circular data
X, y = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Plot side-by-side results
plt.figure(figsize=(10, 4))

# K-Means result
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', s=40)
plt.title("K-Means Clustering on make_circles")

# DBSCAN result
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', s=40)
plt.title("DBSCAN Clustering on make_circles")

plt.tight_layout()
plt.show()
```

**Q44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.**

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import numpy as np

# Load Iris dataset
iris = load_iris()
X = iris.data

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Compute Silhouette Coefficients
sil_values = silhouette_samples(X, labels)
avg_score = silhouette_score(X, labels)

# Plot Silhouette Coefficients for each sample
y_lower = 10
plt.figure(figsize=(8, 5))
for i in range(3):
    cluster_sil_values = sil_values[labels == i]
    cluster_sil_values.sort()
    size_cluster = cluster_sil_values.shape[0]
    y_upper = y_lower + size_cluster
    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, cluster_sil_values)
    plt.text(-0.05, y_lower + 0.5 * size_cluster, str(i))
    y_lower = y_upper + 10  # space between clusters

plt.axvline(x=avg_score, color="red", linestyle="--")
plt.title("Silhouette Plot for Iris Dataset (K-Means)")
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Cluster")
plt.show()
```

**Q45. Generate synthetic data using make_blobs and apply Agglomerative Clustering with 'average' linkage.

**Q46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4
features).**

```python
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Wine dataset
data = load_wine()
X = data.data[:, :4]  # Use first 4 features
feature_names = data.feature_names[:4]

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Create DataFrame for visualization
df = pd.DataFrame(X, columns=feature_names)
df['Cluster'] = labels

# Visualize using Seaborn pairplot
sns.pairplot(df, hue='Cluster', palette='Set2', diag_kind='kde')
plt.suptitle("K-Means Clustering on Wine Dataset (First 4 Features)", y=1.02)
plt.show()
```

**Q47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
count.**

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# Generate noisy blob data
X, y = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=5)
labels = dbscan.fit_predict(X)

# Count clusters (excluding noise)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print("Number of clusters found:", n_clusters)
print("Number of noise points:", n_noise)
```

**Q48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the
clusters.**

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load the Digits dataset
digits = load_digits()
X = digits.data

# Reduce dimensions using t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X)

# Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agg.fit_predict(X_tsne)

# Plot clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10', s=40)
plt.title("Agglomerative Clustering on Digits Dataset (t-SNE Reduced)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```
